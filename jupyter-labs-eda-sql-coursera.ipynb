## Overview of the DataSet

It is the only private company ever to return a spacecraft from low-earth orbit, which it first accomplished in December 2010.
SpaceX advertises Falcon 9 rocket launches on its website with a cost of 62 million dollars wheras other providers cost upward of 165 million dollars each, much of the savings is because Space X can reuse the first stage.

Therefore if we can determine if the first stage will land, we can determine the cost of a launch.

This information can be used if an alternate company wants to bid against SpaceX for a rocket launch.

This dataset includes a record for each payload carried during a SpaceX mission into outer space.

### Task 1

##### Display the names of the unique launch sites  in the space mission


```python
%sql SELECT DISTINCT Launch_Site FROM Spacex
```

```python
print(spacex_df['Launch_Site'].unique())
```

### Task 2

##### Display 5 records where launch sites begin with the string 'CCA'


```python
spacex_df[spacex_df['Launch_Site'].str.startswith('CCA')].head(5)
```

```python
%sql SELECT * FROM Spacex WHERE Launch_Site LIKE 'CCA%' LIMIT 5
```

### Task 3

##### Display the total payload mass carried by boosters launched by NASA (CRS)


```python
print(spacex_df.loc[spacex_df['Customer'] == 'NASA (CRS)', 'PAYLOAD_MASS__KG_'].sum())
```

```python
%sql SELECT SUM(PAYLOAD_MASS__KG_) AS Total_PayloadMass FROM Spacex WHERE Customer LIKE 'NASA (CRS)'
```

### Task 4

##### Display average payload mass carried by booster version F9 v1.1


```python
print(spacex_df.loc[spacex_df['Booster_Version'].str.startswith('F9 v1.1'), 'PAYLOAD_MASS__KG_'].mean())
```

```python
%sql SELECT AVG(PAYLOAD_MASS__KG_) AS Avg_PayloadMass FROM Spacex WHERE Booster_Version = 'F9 v1.1'
```

### Task 5

##### List the date when the first successful landing outcome in ground pad was acheived.

*Hint:Use min function*


```python
spacex_df.loc[spacex_df['Landing _Outcome'] == 'Success (ground pad)', 'Date'].min()
```

```python
%%sql SELECT MIN(DATE(STR_TO_DATE(Date, '%d-%m-%Y'))) AS First_SuccessLanding FROM Spacex 
        WHERE `Landing _Outcome` = 'Success (ground pad)'
```

### Task 6

##### List the names of the boosters which have success in drone ship and have payload mass greater than 4000 but less than 6000


```python
spacex_df.loc[(spacex_df['PAYLOAD_MASS__KG_'] > 4000) & (spacex_df['PAYLOAD_MASS__KG_'] < 6000) & (spacex_df['Landing _Outcome'] == 'Success (drone ship)'), 'Booster_Version']
```

```python
%%sql SELECT Booster_Version From Spacex WHERE `Landing _Outcome` = 'Success (drone ship)'
        AND PAYLOAD_MASS__KG_ > 4000 AND PAYLOAD_MASS__KG_ < 6000
```

### Task 7

##### List the total number of successful and failure mission outcomes


```python
%sql SELECT COUNT(Mission_Outcome) AS MissionSuccess FROM Spacex WHERE Mission_Outcome LIKE 'Success%'
```

```python
%sql SELECT COUNT(Mission_Outcome) AS MissionSuccess FROM Spacex WHERE Mission_Outcome LIKE 'Failure%'
```

### Task 8

##### List the   names of the booster_versions which have carried the maximum payload mass. Use a subquery


```python

%sql select Booster_version from Spacex where PAYLOAD_MASS__KG_ in (select max(PAYLOAD_MASS__KG_) from Spacex) 
```

### Task 9

##### List the failed landing_outcomes in drone ship, their booster versions, and launch site names for in year 2015


```python
spacex_df['Date'] = pd.to_datetime(spacex_df['Date'])
spacex_df.loc[(spacex_df['Date'].dt.year == 2015) & (spacex_df['Landing _Outcome'] == 'Failure (drone ship)'), 'Launch_Site']
```

```python
failedLanding_2015 = spacex_df[['Landing _Outcome', 'Booster_Version', 'Launch_Site', 'Date']]
failedLanding_2015 = failedLanding_2015.loc[(failedLanding_2015['Landing _Outcome'] == 'Failure (drone ship)') & (failedLanding_2015['Date'].dt.year == 2015)]
failedLanding_2015.head()
```

```python
%%sql SELECT Booster_Version, Launch_Site, `Landing _Outcome`, Date FROM Spacex 
        WHERE `Landing _Outcome` = 'Failure (drone ship)' 
        AND STR_TO_DATE(Date, '%d-%m-%Y') BETWEEN '2015-01-01' AND '2015-12-31'
```

### Task 10

##### Rank the count of landing outcomes (such as Failure (drone ship) or Success (ground pad)) between the date 2010-06-04 and 2017-03-20, in descending order


```python
rankOutcome = spacex_df[(spacex_df['Landing _Outcome'] == 'Failure (drone ship)') | (spacex_df['Landing _Outcome'] == 'Success (ground pad)')]
rankOutcome = rankOutcome[(rankOutcome['Date'] >= '2010-06-04') & (rankOutcome['Date'] <= '2017-03-20')]
rankOutcome['rank'] = rankOutcome['Landing _Outcome'].rank(ascending=False)
rankOutcome.head()
```

```python
%%sql SELECT `Landing _Outcome`, COUNT(`Landing _Outcome`) FROM Spacex 
        WHERE STR_TO_DATE(Date, '%d-%m-%Y') BETWEEN '2010-06-04' AND '2017-03-20'
        GROUP BY `Landing _Outcome` 
        ORDER BY COUNT(`Landing _Outcome`) DESC
```
